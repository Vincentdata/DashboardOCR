# Analyse du je de données eurostat : Données mensuelles de demandes d'asile par Nationalité, age et sexe (https://ec.europa.eu/eurostat/web/main/data/database#Data%20navigation%20tree)

Eurostat est une direction générale de la Commission européenne chargée de l'information statistique à l'échelle communautaire. Elle a pour rôle de produire les statistiques officielles de l'Union européenne, principalement en collectant, harmonisant et agrégeant les données publiées par les instituts nationaux de statistiques des pays membres de l'Union européenne, des pays candidats à l'adhésion et des pays de l'Association européenne de libre-échange.

## Import des librairies et du dataset

```python
import pandas as pd
import numpy as np

df = pd.read_csv("estat_migr_asyappctzm.tsv", sep="\t")
df.head()
```

## Préparation des dataframes france et europe

```python
pd.set_option('display.max_rows', 40, 'display.max_columns',230)
```

```python
# Split et suppression de la colonne des caractéristiques concaténées
df[['freq','unit','citizen','sex','applicant','age','geo']] = \
    df['freq,unit,citizen,sex,applicant,age,geo\\TIME_PERIOD'].str.split(',', expand=True)
df = df.drop(columns=['freq,unit,citizen,sex,applicant,age,geo\\TIME_PERIOD'])

df.head()
```

```python
# Nettoyage des noms de colonnes
df.columns = df.columns.str.strip()
```

```python
# Selection des données concernant la nationalité soudanaise
df_sudan = df[df['citizen'] == 'SD']
df_sudan.head()
```

### Variables freq, unit, citizen, sex, applicant, age, geo

**freq**

```python
df_sudan['freq'].unique()
```

frequence temporelle (Monthly)

```python
# Suppression colonne freq
df_sudan = df_sudan.drop('freq', axis=1)
```

**uniq**

```python
df_sudan['unit'].unique()
```

Les valeurs représentes un nombre de personnes

```python
# Suppression colonne unit
df_sudan = df_sudan.drop('unit', axis=1)
```

**citizen**

```python
df_sudan['citizen'].unique()
```

```python
# Suppression colonne citizen
df_sudan = df_sudan.drop('citizen', axis=1)
```

**sex**

```python
df_sudan['sex'].unique()
```

F → Female (femmes) M → Male (hommes) T → Total (F + M) UNK → Unknown (sexe non renseigné)

**applicant**

```python
df_sudan['applicant'].unique()
```

FRST → First applicants (premières demandes)
SSEQ → Subsequent applicants (demandes suivantes / renouvellements)
TOTAL → Total (FRST + SSEQ)

```python
df_sudan.columns
```

```python
# Selection des premieres demandes uniquement
df_sudan = df_sudan[df_sudan['applicant']=='FRST']

# Suppression de la colonne applicant
df_sudan = df_sudan.drop('applicant', axis=1)
```

```python
df_sudan
```

**age**

```python
df_sudan['age'].unique()
```

TOTAL → toutes les classes d’âge confondues UNK → âge inconnu (Unknown) Classes d’âge : Y_LT14 → moins de 14 ans (Youth Less Than 14) Y_LT18 → moins de 18 ans Y14-17 → 14 à 17 ans Y18-34 → 18 à 34 ans Y35-64 → 35 à 64 ans Y_GE65 → 65 ans et plus (Greater or Equal 65)

```python
age_map = {
    "Y_LT14": "<14",
    "Y_LT18": "<18",
    "Y14-17": "14-17",
    "Y18-34": "18-34",
    "Y35-64": "35-64",
    "Y_GE65": "65+",
    "TOTAL": "Total",
    "UNK": "Unknown"
}

df_sudan["age"] = df_sudan["age"].map(age_map)
df_sudan['age'].unique()
```

geo

```python
df_sudan['geo'].unique()
```

```python
# Verification des valeurs de sexe inconnu
df_sudan = df_sudan.copy()
df_sudan.columns = df_sudan.columns.str.strip()
df_filtered = df_sudan[df_sudan["sex"] == "UNK"]
unique_values = pd.unique(df_filtered.iloc[:, :218].values.ravel())

unique_values
```

```python
# préparation des datasets france et europe
df_france = df_sudan[df_sudan['geo'] == 'FR']
df_europe = df_sudan
```

```python
# Suppession de l'année 2008, vide de valeurs
df_france = df_france.drop(df.columns[:12], axis=1)
df_france
```

```python
df_france.info()
```

```python
# Verification des valeurs de sexe inconnu
df_france = df_france.copy()
df_france.columns = df_france.columns.str.strip()

df_filtered = df_france[df_france.iloc[:, 207] == "UNK"]

unique_values = pd.unique(df_filtered.iloc[:, :207].values.ravel())

unique_values
```

```python
# Suppression des lignes "sexe inconnu"
df_france = df_france[df_france["sex"] != "UNK"]
```

```python
df_france
```

```python
df_long = df_france.melt(
    id_vars=['geo','citizen'],
    var_name='year month',
    value_name='value'
)
```

```python
df_long
```

```python
df_long['value'] = df_long['value'].astype(str)

# enlever espaces
df_long['value'] = df_long['value'].str.strip()

# enlever flags Eurostat (e, p, etc.)
df_long['value'] = df_long['value'].str.replace(r'[a-zA-Z]', '', regex=True)


# convertir en float
df_long['value'] = pd.to_numeric(df_long['value'], errors='coerce')
```

```python
df_long['value'].isna().sum()
```

```python
df_long = df_long.dropna(subset=['value'])
```

```python
df_france_final = df_long[df_long["value"]!=0]
df_france_final
```

```python
df_france_final['year month'] = df_france_final['year month'].str.strip()
df_france_final['year month'] = pd.to_datetime(df_france_final['year month'], 
format='%Y-%m')
df_france_final['year month'] = df_france_final['year month'].dt.date



df_france_final
```

```python
key_cols = df_france_final.columns[:3].tolist()
value_col = df_france_final.columns[3]

df_france_final_grouped = df_france_final.groupby(key_cols, as_index=False)[value_col].sum()
```
